```python
X = df.drop(columns=['price'], axis=1)
y = df['price']
```

```python
X.head()
```

```python
y.head()
```

# Two types of encoding :
One hot encoding 

label encoding

# ONE HOT ENCODING

```python
X_one_encoded = pd.get_dummies(X,columns=['model','transmission','fuelType'], drop_first=True)
```

```python
X_one_encoded.head()
```

```python
X_one_encoded = X_one_encoded.astype('int')
X_one_encoded.head()
```

# LABEL ENCODING

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

columns = ['model','transmission','fuelType']
Xlabel_encoded = X
for i in columns:
    Xlabel_encoded[i] = encoder.fit_transform(Xlabel_encoded[i])
```

```python
Xlabel_encoded.head()
```

# STANDARD SCALING FOR BOTH THE ENCODING METHODS

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
numeric_columns = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
X_one_encoded[numeric_columns] = scaler.fit_transform(X_one_encoded[numeric_columns])
X_one_encoded.head()
```

```python
Xlabel_encoded[columns] = scaler.fit_transform(Xlabel_encoded[columns])
Xlabel_encoded.head()
```

# CREATING MODEL FOR BOTH THE ENCODING FOR COMPARISON 

# One-Hot Encoding

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_encoded, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
n = X_train.shape[0]
p = X_train.shape[1]
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
adjusted_r2
```

# Label Encoding

```python
X_train, X_test, y_train, y_test = train_test_split(Xlabel_encoded, y, test_size=0.2, random_state=42)
```

```python
model2 = LinearRegression()
model2.fit(X_train, y_train)
```

```python
y_pred = model2.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
n = X_train.shape[0]
p = X_train.shape[1]
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
adjusted_r2
```
